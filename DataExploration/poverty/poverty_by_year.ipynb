```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
total_crimes = pd.read_csv('data/crimes_by_year.csv', header=None)
bronx_poverty = pd.read_csv('data/bronx_poverty.csv', header=None)
bronx_crimes = pd.read_csv('data/crimes_by_bronx_per_year.csv', header=None)

queens_poverty = pd.read_csv('data/queens_poverty.csv', header=None)
queens_crimes = pd.read_csv('data/crimes_by_queens_per_year.csv', header=None)

manhattan_poverty = pd.read_csv('data/manhattan_poverty.csv', header=None)
manhattan_crimes = pd.read_csv('data/crimes_by_manhattan_per_year.csv', header=None)

brooklyn_poverty = pd.read_csv('data/brooklyn_poverty.csv', header=None)
brooklyn_crimes = pd.read_csv('data/crimes_by_brooklyn_per_year.csv', header=None)

statenisland_poverty = pd.read_csv('data/statenisland_poverty.csv', header=None)
statenisland_crimes = pd.read_csv('data/crimes_by_statenisland_per_year.csv', header=None)
```

```python
fig,ax1 = plt.subplots()
plt.title('Bronx: Poverty rate vs Number of crimes')
ax1.plot(bronx_poverty[0], bronx_poverty[1], 'b-', label="Poverty")
ax1.set_xlabel('year')
ax1.set_ylabel('Poverty rate', color='b')
ax1.tick_params('y', colors='b')

ax2 = ax1.twinx()
ax2.plot(bronx_poverty[0], bronx_crimes[1][:-1], 'r-', label="Number of crimes")
ax2.set_ylabel('Number of crimes', color='r')
ax2.tick_params('y', colors='r')

fig.tight_layout()
plt.show()
```

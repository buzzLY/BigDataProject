```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
temperature = pd.read_csv('data/ny_temperature_month.csv', header=None)
crimes = pd.read_csv('data/crimes_by_month.csv', header=None)
```

```python
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
month_temp_series = pd.Series(temperature[1].values, index=month_order)
month_crimes_series = pd.Series(crimes[1].values, index=month_order)
month_temp = pd.DataFrame({'month':month_temp_series.index, 'value':month_temp_series.values})
month_crimes = pd.DataFrame({'month':month_crimes_series.index, 'value':month_crimes_series.values})
```

```python
fig,ax1 = plt.subplots()
plt.title('Crime rate vs Temperature')
ax1.plot(month_temp['value'], 'b-', label="Average Temperature")
ax1.set_xlabel('Month')
ax1.set_ylabel('Average Temperature', color='b')
ax1.tick_params('y', colors='b')

ax2 = ax1.twinx()
ax2.plot(month_crimes['value'], 'r-', label="Number of crimes")
ax2.set_ylabel('Number of crimes', color='r')
ax2.tick_params('y', colors='r')
plt.xticks(np.arange(12), month_order)
ax1.tick_params(axis='x', rotation=90)
ax2.tick_params(axis='x', rotation=90)
plt.show()
```

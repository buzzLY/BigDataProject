```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
borough = pd.read_csv('data/crimes_borough.csv', header=None)
law_cat = pd.read_csv('data/crimes_law_cat.csv', header=None)
crm_atpt_cptd = pd.read_csv('data/crimes_A_C.csv', header=None)
loc_of_occur = pd.read_csv('data/crimes_loc_occ.csv', header=None)
```

```python
plt.bar(borough[0], borough[1],color='red')
plt.xlabel("Borough")
plt.ylabel("Number of Crimes")
plt.title("Crimes by Borough")
plt.show()
```

```python
plt.bar(law_cat[0], law_cat[1], color='orange')
plt.xlabel("Offense")
plt.ylabel("Number of Crimes")
plt.title("Crimes by Level of Offense")
plt.show()
```

```python
plt.bar(loc_of_occur[0], loc_of_occur[1],color='blue')
plt.xlabel("Location of Occurence")
plt.ylabel("Number of Crimes")
plt.title("Crimes by Location of Occurence")
plt.show()
```

```python
plt.bar(crm_atpt_cptd[0], crm_atpt_cptd[1], color='green')
plt.xlabel("Attempted or Completed")
plt.ylabel("Number of Crimes")
plt.title("Crimes by Status")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
unsafe = pd.read_csv('data/parks/unsafe_parks.csv', header=None)
bronx_unsafe = pd.read_csv('data/parks/bronx_unsafe_parks.csv', header=None)
brooklyn_unsafe = pd.read_csv('data/parks/brooklyn_unsafe_parks.csv', header=None)
queens_unsafe = pd.read_csv('data/parks/queens_unsafe_parks.csv', header=None)
manhattan_unsafe = pd.read_csv('data/parks/manhattan_unsafe_parks.csv', header=None)
staten_unsafe = pd.read_csv('data/parks/statenisland_unsafe_parks.csv', header=None)
```

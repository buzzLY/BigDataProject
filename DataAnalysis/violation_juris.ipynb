```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
violation_data = pd.read_csv('data/crime_violation_juris.csv', header=None)
```

```python
plt.bar(range(len(violation_data[2])), violation_data[0].astype(float))
plt.xticks(range(len(violation_data[2])), violation_data[2], rotation=90)
plt.xlabel("Jurisdiction of Violations")
plt.ylabel("Number of Violations")
plt.title("Violations by Jurisdiction")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
ofns1 = pd.read_csv('data/ofns/ofns_by_year_1.csv', header=None)
ofns2 = pd.read_csv('data/ofns/ofns_by_year_2.csv', header=None)
ofns3 = pd.read_csv('data/ofns/ofns_by_year_3.csv', header=None)
ofns4 = pd.read_csv('data/ofns/ofns_by_year_4.csv', header=None)
ofns5 = pd.read_csv('data/ofns/ofns_by_year_5.csv', header=None)
ofns_all = [ofns1, ofns2, ofns3, ofns4, ofns5]
```

```python
for i in range(0,5):
    plt.plot(ofns_all[i][0], ofns_all[i][1], label=ofns_all[i][2][0])
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title("Top 5 Ofenses By Year")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('data/crimes_in_BRONX_by_LAW_CD.csv', header=None)
```

```python
plt.bar(data[0], data[1].astype(float))
#plt.plot(data[0], data[1], "b--", marker=".")
plt.xlabel("Crime Type")
plt.ylabel("Number of Crimes")
plt.title("No Of Crime Types in Bronx")
plt.show()
```

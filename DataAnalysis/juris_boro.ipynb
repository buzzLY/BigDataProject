```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
bronx_data = pd.read_csv('data/juris_boro/juris_bronx_boro.csv', header=None)
brooklyn_data = pd.read_csv('data/juris_boro/juris_brooklyn_boro.csv', header=None)
queens_data = pd.read_csv('data/juris_boro/juris_queens_boro.csv', header=None)
manhattan_data = pd.read_csv('data/juris_boro/juris_manhattan_boro.csv', header=None)
staten_data = pd.read_csv('data/juris_boro/juris_statenisland_boro.csv', header=None)
```

```python
plt.plot(range(len(bronx_data[1])), bronx_data[0].astype(float), label="Bronx")
plt.xticks(range(len(bronx_data[1])), bronx_data[1], rotation=90)
plt.xlabel("Jurisdiction of Crimes")
plt.ylabel("Number of Crimes")
plt.title("Crimes by Jurisdiction in Bronx")
plt.show()
```

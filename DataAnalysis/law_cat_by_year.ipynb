```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
felony_data = pd.read_csv('data/law_cat/felony_by_year.csv', header=None)
violation_data = pd.read_csv('data/law_cat/violation_by_year.csv', header=None)
misdemeanor_data = pd.read_csv('data/law_cat/misdemeanor_by_year.csv', header=None)
```

```python
plt.plot(felony_data[0], felony_data[1], label="FELONY")
plt.plot(violation_data[0], violation_data[1], label="VIOLATION")
plt.plot(misdemeanor_data[0], misdemeanor_data[1], label="MISDEMEANOR")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

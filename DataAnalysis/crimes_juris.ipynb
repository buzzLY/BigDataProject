```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('data/juris.csv', header=None)
```

```python
plt.bar(range(len(data[1])), data[0].astype(float))
plt.xticks(range(len(data[1])), data[1], rotation=90)
plt.xlabel("Jurisdiction of Crime")
plt.ylabel("Number of Crimes")
plt.title("Crimes by Jurisdiction")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
bronx_data = pd.read_csv('data/prec_boro/prec_bronx_boro.csv', header=None)
brooklyn_data = pd.read_csv('data/prec_boro/prec_brooklyn_boro.csv', header=None)
queens_data = pd.read_csv('data/prec_boro/prec_queens_boro.csv', header=None)
manhattan_data = pd.read_csv('data/prec_boro/prec_manhattan_boro.csv', header=None)
staten_data = pd.read_csv('data/prec_boro/prec_statenisland_boro.csv', header=None)
```

```python
plt.plot(range(len(bronx_data[2])), bronx_data[0].astype(float), label="Bronx")
plt.xticks(range(len(bronx_data[2])), bronx_data[2])
plt.xlabel("Precincts")
plt.ylabel("Number of Crimes")
plt.title("Crimes by Precinct in Bronx")
plt.show()
```

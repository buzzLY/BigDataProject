```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
safe = pd.read_csv('data/parks/safe_parks.csv', header=None)
bronx_safe = pd.read_csv('data/parks/bronx_safe_parks.csv', header=None)
brooklyn_safe = pd.read_csv('data/parks/brooklyn_safe_parks.csv', header=None)
queens_safe = pd.read_csv('data/parks/queens_safe_parks.csv', header=None)
manhattan_safe = pd.read_csv('data/parks/manhattan_safe_parks.csv', header=None)
staten_safe = pd.read_csv('data/parks/statenisland_safe_parks.csv', header=None)
```

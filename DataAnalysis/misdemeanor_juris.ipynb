```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
misdemeanor_data = pd.read_csv('data/crime_misdemeanor_juris.csv', header=None)
```

```python
plt.bar(range(len(misdemeanor_data[2])), misdemeanor_data[0].astype(float))
plt.xticks(range(len(misdemeanor_data[2])), misdemeanor_data[2], rotation=90)
plt.xlabel("Jurisdiction of Misdemeanor")
plt.ylabel("Number of Misdemeanor")
plt.title("Misdemeanor by Jurisdiction")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
felony_data = pd.read_csv('data/crime_felony_juris.csv', header=None)
```

```python
plt.bar(range(len(felony_data[2])), felony_data[0].astype(float))
plt.xticks(range(len(felony_data[2])), felony_data[2], rotation=90)
plt.xlabel("Jurisdiction of Felony")
plt.ylabel("Number of Felonies")
plt.title("Felonies by Jurisdiction")
plt.show()
```
